```python
import gym

from stable_baselines.common.policies import MlpPolicy
from stable_baselines.common import make_vec_env
from stable_baselines import PPO2
```

```python
env0 = gym.make('CartPole-v1')
print(env0.action_space)
print(env0.observation_space)
```

```python
print(env0.observation_space.high)
print(env0.observation_space.low)
```

```python
# multiprocess environment
env = make_vec_env('CartPole-v1', n_envs=8)

model = PPO2(MlpPolicy, env, verbose=1)
model.learn(total_timesteps=200000)
model.save("ppo2_cartpole")
```

```python
# Enjoy trained agent
env2 = gym.make('CartPole-v1')
obs = env2.reset()
done = False

while not done:
    action, _states = model.predict(obs)
    obs, rewards, done, info = env2.step(action)
    env2.render()

env2.close()
```
