```python
import gym

from stable_baselines.common.policies import MlpPolicy, CnnPolicy, CnnLstmPolicy
from stable_baselines.common import make_vec_env
from stable_baselines import PPO2
```

```python
agent='CartPole-v1'
#agent='MountainCar-v0'
#agent='Pendulum-v0'
```

```python
env0 = gym.make(agent)
print(env0.action_space)
print(env0.observation_space)
```

```python
# multiprocess environment
paralelo=8
env = make_vec_env(agent, n_envs=paralelo)

model = PPO2(MlpPolicy, env, verbose=1, n_steps=int(512/paralelo), cliprange_vf=-1, nminibatches=paralelo)
model.learn(total_timesteps=50000)
model.save("ppo2_agent")
```

```python
# Enjoy trained agent
env2 = gym.make(agent)
obs = env2.reset()
done = False

while not done:
    action, _states = model.predict(obs)
    obs, rewards, done, info = env2.step(action)
    env2.render()

env2.close()
```
